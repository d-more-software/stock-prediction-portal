```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Données du tableau
t = [0.00, 1.00, 2.00, 3.00, 4.00, 5.00]
delta_t = [0.00, 0.005, 0.010, 0.015, 0.020, 0.025]
X = [0.25, 4.10, 8.20, 12.00, 16.30, 20.90]
delta_X = [0.05, 0.05, 0.05, 0.05, 0.05, 0.05]
```

```python

# Création du graphique
plt.figure(figsize=(10, 6))
plt.errorbar(t, X, xerr=delta_t, yerr=delta_X, 
             fmt='o-', color='blue', markersize=6, 
             capsize=4, capthick=1, elinewidth=1,
             label='Données expérimentales')

# Personnalisation du graphique
plt.title('Graphique Position vs Temps', fontsize=14, fontweight='bold')
plt.xlabel('Temps $t$ (s)', fontsize=12)
plt.ylabel('Position $X$ (2m)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()

# Ajustement des limites des axes
plt.xlim(0, 5.5)
plt.ylim(0, 22)
```
