## Subplots and Legends

```python
days = [1,2,3,4,5]
original_price = [10,20,25,30,40]
predicted_price = [8, 21, 25, 28, 38]
```

```python

plt.plot(days, predicted_price,'r-o', label='predicted_price')
plt.plot(days, original_price,'b-o', label='original_price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')
```

```python
plt.plot(days,predicted_price)
plt.title('Plot with legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')
```

# Subplots

```python
# First subplot - original_price
plt.subplot(3,1,1)
plt.plot(days, original_price,'b-o', label='original_price')
plt.legend()
plt.title('Original prices over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')

# Second subplot - original_price
plt.subplot(3,1,2)
plt.plot(days, predicted_price,'r-o', label='predicted_price')
plt.legend()
plt.title('predicted prices over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')

# third subplot - original_price
price_difference = [o - p for o, p in zip(original_price,predicted_price) ]
plt.subplot(3,1,3)
plt.bar(days, price_difference,color='purple', label='Price Difference')
plt.legend()
plt.title('predicted prices over time')
plt.xlabel('Days')
plt.ylabel('Price')

plt.tight_layout()
```

## Saving the figure

```python

plt.plot(days, predicted_price,'r-o', label='predicted_price')
plt.plot(days, original_price,'b-o', label='original_price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')

plt.savefig('example_prediction.png',dpi=300)
```
